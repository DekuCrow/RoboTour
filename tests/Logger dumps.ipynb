```python
from __future__ import print_function

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

sns.set_palette('dark')
```

```python
def logger_plot(filename, title, channels):
    data = pd.read_csv(filename)
    
    
    t0 = data['timestamp'][0]
    data['timestamp'] -= t0
    
    data = data.drop([0])
    
    time = data['timestamp']
    fig = plt.figure(figsize = (12, 8))
    plt.ylim(0,255)
    

    col = 0
    
    for s in channels:
        
        plt.plot(time, data[s], '-', color="C{}".format(col%10), alpha=0.75, label=s)
        col += 1
    
    
    
    plt.title(title, fontsize=15)
    plt.xlabel("Time", fontsize=13)
    plt.ylabel("Sensor reading", fontsize=13)
    
    plt.grid(linestyle='--')
    plt.legend(fontsize=15)
    plt.show()
    
    fig.savefig(filename[:-4]+".png", dpi=fig.dpi)
```

```python
logger_plot("./logs/line_sensor_test_03_03_22-35.csv", "Sensor value and threshold", ['l0', 'l0th', 'l2', 'l2th', 'output'])
```
