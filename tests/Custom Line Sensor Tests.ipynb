```python

def sensor_plot(filename, title, sensors):
    data = pd.read_csv(filename)
    
    
    t0 = data['timestamp'][0]
    data['timestamp'] -= t0
    
    data = data.drop([0])
    
    time = data['timestamp']
    fig = plt.figure(figsize = (12, 8))
    plt.ylim(0,255)
    

    for s in sensors:
        plt.plot(time, data[s], '-', color=c[s], alpha=0.75, label=l[s])

    
    
    
    plt.title(title, fontsize=15)
    plt.xlabel("Time", fontsize=13)
    plt.ylabel("Sensor reading", fontsize=13)
    
    plt.grid(linestyle='--')
    plt.legend(fontsize=15)
    plt.show()
    
    fig.savefig(filename[:-4]+".png", dpi=fig.dpi)
```

```python
filename = "./logs/line_sensor_test_27_02_20-07.csv"
title = "Perpendicular pass over the line"
data = pd.read_csv(filename)


t0 = data['timestamp'][0]
data['timestamp'] -= t0

data = data[data.timestamp > 10.5]
data = data[data.timestamp < 13]


time = data['timestamp']
fig = plt.figure(figsize = (12, 8))
plt.ylim(150,255)

time = (time-10.5) * 100 / 360 * 8.18 * np.pi


for s in sensors:
    plt.plot(time, data[s], 'o-', color=c[s], alpha=0.75, label=l[s])




plt.title(title, fontsize=15)
plt.xlabel("Distance [cm]", fontsize=13)
plt.ylabel("Sensor reading", fontsize=13)

plt.grid(linestyle='--')
plt.legend(fontsize=15)
plt.show()

fig.savefig(title+".png", dpi=fig.dpi)
```
